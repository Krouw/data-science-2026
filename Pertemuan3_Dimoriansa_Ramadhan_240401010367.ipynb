```python
import pandas as pd
import numpy as np
import requests
from scipy.stats.mstats import winsorize
```

```python
from google import colab
df = pd.read_csv('/housing_dirty.csv')
print("Dataset berhasil dimuat!")
```

```python
#Eksplorasi awal

print("\n== Informasi Dataset ==")
df.info()

print("\n== Deskripsi Statistik ==")
df.describe()

print("\n== Jumlah Missing Values per Kolom ==")
df.isnull().sum()
```

```python
#Hapus baris duplikat

df.drop_duplicates(inplace=True)
print("Setelah hapus duplikat:", df.shape)
```

```python
#Normalisasi string

df['kota'] = df['kota'].str.strip().str.title()
df['kondisi'] = df['kondisi'].str.strip().str.lower()

print(df.head())
```

```python
#Imputasi missing values

df['luas_m2'] = df['luas_m2'].fillna(df['luas_m2'].median())
df['harga_juta'] = df['harga_juta'].fillna(df['harga_juta'].median())
df['kamar'] = df['kamar'].fillna(df['kamar'].mode()[0])

print(df.head())
print(df.isnull().sum())
```

```python
#Tangani Outlier (IQR Fence)

print("\n== hasil tangani outlier ==\n")

for col in ['harga_juta', 'luas_m2', 'tahun_bangun']:
    Q1, Q3 = df[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    df[col] = df[col].clip(Q1-1.5*IQR, Q3+1.5*IQR)

print(df[['harga_juta', 'luas_m2', 'tahun_bangun']].describe())
```

```python
#Validasi & Ekspor

assert df.isnull().sum().sum() == 0, 'Masih ada missing!'
assert df.duplicated().sum() == 0, 'Masih ada duplikat!'


print("\n== Validasi berhasil ==\n")
print("Dataset sudah bersih")
print('Shape akhir:', df.shape)
```

```python
df.to_csv('housing_clean.csv', index=False)

print("\n== Ekspor berhasil ==\n")
print("Dataset telah disimpan menjadi 'housing_clean.csv'")
```

```python
import requests
import pandas as pd
from pandas import json_normalize
import time

print("\n== Akses API JSONPlaceholder ==\n")

# URL API
url = 'https://jsonplaceholder.typicode.com/posts'

try:
    # Request API dengan timeout
    response = requests.get(URL, timeout=10)

    # Cek status code
    if response.status_code == 200:

        # Convert JSON ke Python
        data = response.json()

        # Ubah menjadi DataFrame
        df = json_normalize(data, sep='_')

        print("Request berhasil\n")

        # Tampilkan beberapa kolom
        print(df.head())

    else:
        print(f'Error Status Code: {response.status_code}')

# Tangani error koneksi
except requests.exceptions.ConnectionError:
    print("Koneksi gagal ke API")

# Tangani timeout
except requests.exceptions.Timeout:
    print("Request timeout")

# Tangani error lainnya
except requests.exceptions.RequestException as e:
    print(f"Terjadi error: {e}")
```
